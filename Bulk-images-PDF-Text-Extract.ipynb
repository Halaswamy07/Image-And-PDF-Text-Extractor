```python
import os

# Function definitions as you've provided
import pytesseract
from PIL import Image
import pdf2image

# Set the tesseract executable path
pytesseract.pytesseract.tesseract_cmd = r'C:\Program Files\Tesseract-OCR\tesseract.exe'
# Set the path to the Poppler bin directory
poppler_path = r'C:\Program Files\poppler-24.02.0\Library\bin'

def extract_text_from_image(image_path):
    try:
        image = Image.open(image_path)
        text = pytesseract.image_to_string(image)
        return text
    except Exception as e:
        print(f"An error occurred while processing the image: {e}")
        return None

def extract_text_from_pdf(pdf_path):
    try:
        pages = pdf2image.convert_from_path(pdf_path, poppler_path=poppler_path)
        text = ""
        for page in pages:
            text += pytesseract.image_to_string(page)
        return text
    except Exception as e:
        print(f"An error occurred while processing the PDF: {e}")
        return None

# Function to process all images in a directory
def process_images_in_directory(directory):
    for filename in os.listdir(directory):
        if filename.endswith(".jpg") or filename.endswith(".png"):
            image_path = os.path.join(directory, filename)
            print(f"Extracting text from image: {filename}")
            image_text = extract_text_from_image(image_path)
            if image_text:
                print("Extracted text:", image_text)
            else:
                print("No text extracted.")

# Function to process all PDFs in a directory
def process_pdfs_in_directory(directory):
    for filename in os.listdir(directory):
        if filename.endswith(".pdf"):
            pdf_path = os.path.join(directory, filename)
            print(f"Extracting text from PDF: {filename}")
            pdf_text = extract_text_from_pdf(pdf_path)
            if pdf_text:
                print("Extracted text:", pdf_text)
            else:
                print("No text extracted.")

# Example usage:
image_directory = r'C:\Users\halas\my project\image'
pdf_directory = r'C:\Users\halas\my project\document'

# Process all images in the image directory
print("Processing images...")
process_images_in_directory(image_directory)

# Process all PDFs in the PDF directory
print("\nProcessing PDFs...")
process_pdfs_in_directory(pdf_directory)
```
